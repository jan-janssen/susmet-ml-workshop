```python
import matplotlib.pyplot as plt
import json
```

```python
folder = "data/"
filenames = [
    "1-s2.0-S0010938X17320796-main.mmd_t00_mistral.json",
    "1-s2.0-S0010938X18323187-main.mmd_t00_mistral.json",
    "1-s2.0-S1359645415003547-main.mmd_t00_mistral.json"
            ]
```

```python
data = []
for filename in filenames:
    f = open(folder + filename)
    tmp = json.load(f)
    data.append(tmp)
```

```python
print('nominal_composition' in data[0])
```

```python
comp_0 = data[0]["experiment"]['materials']['compositions'][0]['nominal_composition']
comp_1 = data[1]["experiment"]['materials']['compositions'][0]['nominal_composition']
comp_2 = data[2]["experiment"]['materials']['compositions'][0]['nominal_composition']

names = [comp_0, comp_1, comp_2]
```

```python
temp_0 = data[0]["experiment"]['materials']['compositions'][0]['processing_conditions']['heat_treatment']['homogenization']['temperature']['value']
temp_1 = data[1]["experiment"]['materials']['compositions'][0]['processing_conditions']['heat_treatment']['homogenization']['temperature']['value']
temp_2 = data[1]["experiment"]['materials']['compositions'][0]['processing_conditions']['heat_treatment']['homogenization']['temperature']['value']

temperatures = [temp_0, temp_1, temp_2]
```
